# NBA player similarity using regular and advanced stats #

Using advanced statistics from the 2012-2013 NBA season (www.basketball-reference.com) and the affinity propagation algorithm described by Dueck (2009), players can be categorized into groups and described by examplars, ie. the player that best represents the group. 

Regular stats are per 36 minute totals and percentages for rebounding, shooting, and points, whereas advanced stats include true and effective percentages, rebound, assist, and turnover percentage, and free throw rate.

The analysis is filtered by players with over 800 total minutes throught the season, or about 10 minutes per game, if healthy.

```python
import os
import csv
import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from pylab import *
%matplotlib inline

# regular stats
df_per36 = pd.read_csv(os.getcwd() + '/Data/nba/nba1213per36.csv', index_col = 1).fillna(0)
cols = [6,10,12,13,15,16,18,19,20,22,23,24,25,26,27] # stats of interest
df = df_per36.ix[df_per36.MP>800,cols] # filter by minutes per game
# df_copy = df_per36.ix[df_per36.MP>800,cols] # copy

# advanced stats
df_adv = pd.read_csv(os.getcwd() + '/Data/nba/nba1213adv.csv', index_col = 1).fillna(0)
cols2 =[7,8,9,10,11,12,14,15,16,17,18] # stats of interest
df2 = df_adv.ix[df_adv.MP>800,cols2] # filter by minutes per game
# df2_copy = df_adv.ix[df_adv.MP>800,cols2] # copy

# summary of both stats
# print df.describe()
```

```python
# go through the stats and plot histograms for each regular stat
for i in df.columns: 
    plt.figure()
    plt.hist(df[i].values,20)
    plt.title(i)
plt.show()
```

```python
# go through the stats and plot histograms for each advanced stat
for i in df2.columns: 
    plt.figure()
    plt.hist(df2[i].values,20)
    plt.title(i)
plt.show()
```

## Affinity propagation on regular stats

```python
# run the affinity propagation algorithm on the regular stats
X = StandardScaler().fit_transform(df)
af = AffinityPropagation(preference=None).fit(X)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
df['label'] = labels
```

```python
# # print the examplars
# for i in cluster_centers_indices:
#     pprint.pprint(df_adv.ix[i,:])

for i in set(labels):
    print '\n-------------\n'
    pprint.pprint(i)
    print df.ix[df.label==i,:].index.values
```
